```python
df=pd.read_excel('Sample - Superstore.xls')
df.head()
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['Profit'],color='yellow',rwidth=0.9)
plt.title('Distribution of Profit ',fontsize=16)
plt.xticks(rotation=90)
plt.xlabel('Profit')
plt.ylabel('Distribution of Profit')
plt.show()
```

```python
plt.scatter(df['Quantity'],df['Sales'])
plt.title('Quantity v/s Sales')
plt.show()
```

```python
df1=df[['Quantity','Sales','Profit']]
sns.heatmap(df1.corr(),annot= True)
```

```python
sns.pairplot(df1)
```

```python
plt.bar(df['Country'],df['Sales'])
plt.title('Country and sales')
plt.ylabel('Sales')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.show()
```
